```python
from collections import defaultdict

def find_all_paths(variables):
    # Step 1: Create a directed graph
    graph = defaultdict(list)
    n = len(variables)
    
    for i in range(n):
        for j in range(i + 1, n):
            graph[i].append(j)
    
    # Step 2: DFS to find all paths from a given node
    def dfs(node, path):
        paths = []
        for neighbor in graph[node]:
            paths.append(path + [neighbor])  # Add direct path
            paths.extend(dfs(neighbor, path + [neighbor]))  # Recursive paths
        return paths
    
    # Step 3: Find all paths starting from each node
    all_paths = []
    for i in range(n):
        all_paths.extend(dfs(i, [i]))
    
    # Step 4: Convert indices to variable names
    named_paths = [[variables[idx] for idx in path] for path in all_paths]
    return named_paths

# Example variables
variables = ['first_price', 'offer_counteroffer_diff', 'negotiation_length', 'last_price', 'G']

# Find all paths
paths = find_all_paths(variables)

# Print the results
print("All Possible Paths:")
for path in paths:
    print(path)
```

```python
def indirect_dep(a,b,paths,direct):
    a_b_ind_lis =[]
    added = 0
    for path in paths:
        
        if len(path)>2:
            
            if path[0] == a and path[-1] ==b:
                a_b_ind = 1
                for i,j in zip(path[:-1],path[1:]):
                    added = 1
                    
                    a_b_ind *=direct[(i,j)]
        
                a_b_ind_lis.append(a_b_ind)
    if added:
        return sum(a_b_ind_lis)
    else:
        return 0
                    
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from arch.bootstrap import IIDBootstrap

# Example data (serial mediation)
np.random.seed(42)



#['Control', 'T1_SV_BD', 'T2_SD_BV', 'T3_SV_BA', 'T4_SA_BV',
#       'first_price', 'offer_counteroffer_diff', 'negotiation_length',
#       'last_price', 'start_finish_diff', 'U_s', 'U_b', 'G']


#data = pd.DataFrame({'a': a, 'd': d, 'e': e, 'f': f, 'Y': Y})

def calculate_serial_indirect_effects(data, a = 'T1_SV_BD', Y = 'U_s'#, 'U_b', 'G'
                                     ):
    # Model 1: d ~ a

    d, e,f,g = 'first_price', 'offer_counteroffer_diff', 'negotiation_length', 'last_price'

    #print(data[a].mean())

    # T ->ALL ------------------------
    
    # T -> first price
    model_ad = LinearRegression().fit(data[[a]], data[d])
    ad_a = model_ad.coef_[0]

    # T -> OCD
    model_ae = LinearRegression().fit(data[[a]], data[e])
    ae_a = model_ae.coef_[0]

    
    # D -> negotiation_length
    #model_af = LinearRegression().fit(data[[a]], data[f])
    #af_a = model_af.coef_[0]

    # D -> LP
    #model_ag = LinearRegression().fit(data[[a]], data[g])
    #ag_a = model_ag.coef_[0]
    
    # T -> Y 
    model_aY = LinearRegression().fit(data[[a]], data[Y])
    aY_a = model_aY.coef_[0]

    





    
    # A D ->ALL ------------------------

    # T + FO ~ OFD
    model_ade = LinearRegression().fit(data[[a,d]], data[e])
    ade_a = model_ade.coef_[0]
    ade_d = model_ade.coef_[1]

    # T + FO ~ Y
    model_adY = LinearRegression().fit(data[[a,d]], data[Y])
    adY_a = model_adY.coef_[0]
    adY_d = model_adY.coef_[1]


    # A D E ->ALL ------------------------

    # T + FO + OCD ~ negotiation_length
    model_adef = LinearRegression().fit(data[[a,d,e ]], data[f])
    adef_a = model_adef.coef_[0]
    adef_d = model_adef.coef_[1]
    adef_e = model_adef.coef_[2]

    # T + FO + OCD ~ Y
    model_adeY = LinearRegression().fit(data[[a,d,e]], data[Y])
    adeY_a = model_adeY.coef_[0]
    adeY_d = model_adeY.coef_[1]  
    adeY_e = model_adeY.coef_[2]  
    

    # A D E F->ALL ------------------------

    # T + FO + OCD + negotiation_length ~ last_price
    model_adefg = LinearRegression().fit(data[[a,d,e,f]], data[g])
    adefg_a = model_adefg.coef_[0]
    adefg_d = model_adefg.coef_[1]
    adefg_e = model_adefg.coef_[2]
    adefg_f = model_adefg.coef_[3]

    # T + FO + OCD + negotiation_length  ~ Y
    model_adefY = LinearRegression().fit(data[[a,d,e,f]], data[Y])
    adefY_a = model_adefY.coef_[0]
    adefY_d = model_adefY.coef_[1]  
    adefY_e = model_adefY.coef_[2] 
    adefY_f = model_adefY.coef_[3] 


    # A D E F G->ALL ------------------------

    # T + FO + OCD + negotiation_length + last_price ~ Y
    model_adefgY = LinearRegression().fit(data[[a,d,e,f,g]], data[Y])
    adefgY_a = model_adefgY.coef_[0]
    adefgY_d = model_adefgY.coef_[1]  
    adefgY_e = model_adefgY.coef_[2] 
    adefgY_f = model_adefgY.coef_[3] 
    adefgY_g = model_adefgY.coef_[4] 
        

    # Calculate specific indirect effects
    indirect_a_d_Y = ad_a * adY_d
    indirect_a_e_Y = ae_a * adeY_e
    indirect_d_e_Y = ade_d * adeY_e # e ~ a +d , Y ~ a + d + e

    indirect_e_f_Y = adef_e * adefY_f
    indirect_f_g_Y = adefg_f * adefgY_g

    indirect_a_g_Y = adefg_a*adefgY_g
    indirect_d_g_Y = adefg_d*adefgY_g
    indirect_e_g_Y = adefg_e*adefgY_g




    
    return np.array([

    indirect_a_d_Y
    ,indirect_a_e_Y
    ,indirect_d_e_Y
    ,indirect_e_f_Y
    ,indirect_f_g_Y
        
    ,indirect_a_g_Y
    ,indirect_d_g_Y
    ,indirect_e_g_Y
        
        
    ]) #Return as a vector


# Bootstrapping

results_dict = dict({})

for a in ['T1_SV_BD','T2_SD_BV', 'T3_SV_BA', 'T4_SA_BV']:
    no_chosen_columns = list({ 'T1_SV_BD','T2_SD_BV', 'T3_SV_BA', 'T4_SA_BV'} - {a})
        
    corre_df_2 = corre_df[ corre_df[no_chosen_columns].sum(axis =1) == 0]
    #rune_1_results =  run_reg( corre_df_2, y = 'G',a = 'T1_SV_BD')
    
    for Y in ['U_b','U_s','G']:


        
        print('target',Y, ' treatment',a, '--------------------' )
        
        
        
        
        bs = IIDBootstrap(corre_df_2)
        bs.seed(123)
        ci = bs.conf_int(calculate_serial_indirect_effects, reps=1000, method='bca', size=0.95, extra_kwargs =dict(a=a,Y=Y))
        
        # Get point estimates from the original data
        point_estimates = calculate_serial_indirect_effects(corre_df_2)

        results_dict[(Y,a)] = {'ci':ci, 'point_estimates':point_estimates,
                              'main_effects' : run_reg( corre_df_2, y = Y,a = a)
                              }
        
        # Print results (with interpretation)
        indirect_effect_names = [
            "a -> d -> Y",
            "a -> e -> Y",
            "d -> e -> Y",
            "e -> f -> Y",
            "f -> g -> Y",

            "a -> g -> Y",
            "d -> g -> Y",
            "e -> g -> Y",
                        
        ]
        
        for i, name in enumerate(indirect_effect_names):
            print(f"Indirect Effect: {name}")
            print(f"  Point Estimate: {point_estimates[i]:.3f}")
            print(f"  95% CI: ({ci[0, i]:.3f}, {ci[1, i]:.3f})")
            if ci[0, i] <= 0 <= ci[1, i]:
                print("  NOT Statistically Significant")
            else:
                print("  Statistically Significant")
            print("-" * 30)
# a is the treatement indicator, Y is the as output.
#d, e,f,g = 'first_price', 'offer_counteroffer_diff', 'negotiation_length', 'last_price'
```
